```python
#Correcting typo in age. Example : old string [0-10) to new string (0-10)

for i in range(0,df['age'].shape[0]):
    if df['age'][i] == '[0-10)':
        df['age'] = df['age'].replace(df['age'][i],'(0-10)')
    elif df['age'][i] == '[10-20)':
        df['age'] = df['age'].replace(df['age'][i],'(10-20)')
    elif df['age'][i] == '[20-30)':
        df['age'] = df['age'].replace(df['age'][i],'(20-30)')
    elif df['age'][i] == '[30-40)':
        df['age'] = df['age'].replace(df['age'][i],'(30-40)')
    elif df['age'][i] == '[40-50)':
        df['age'] = df['age'].replace(df['age'][i],'(40-50)')
    elif df['age'][i] == '[50-60)':
        df['age'] = df['age'].replace(df['age'][i],'(50-60)')
    elif df['age'][i] == '[60-70)':
        df['age'] = df['age'].replace(df['age'][i],'(60-70)')
    elif df['age'][i] == '[70-80)':
        df['age'] = df['age'].replace(df['age'][i],'(70-80)')
    elif df['age'][i] == '[80-90)':
        df['age'] = df['age'].replace(df['age'][i],'(80-90)')
    elif df['age'][i] == '[90-100)':
        df['age'] = df['age'].replace(df['age'][i],'(90-100)')
    
    else:
        pass   
```

```python
df['weight'].unique()
```

```python
#Correcting typo in weight. Example : old string [75-100) to new string (75-100)

for i in range(0,df['weight'].shape[0]):
    if df['weight'][i] == '':
        df=df
    elif df['weight'][i] == '[75-100)':
        df['weight'] = df['weight'].replace(df['weight'][i],'(75-100)')
    elif df['weight'][i] == '[50-75)':
        df['weight'] = df['weight'].replace(df['weight'][i],'(50-75)')
    elif df['weight'][i] == '[0-25)':
        df['weight'] = df['weight'].replace(df['weight'][i],'(0-25)')
    elif df['weight'][i] == '[100-125)':
        df['weight'] = df['weight'].replace(df['weight'][i],'(100-125)')
    elif df['weight'][i] == '[25-50)':
        df['weight'] = df['weight'].replace(df['weight'][i],'(25-50)')
    elif df['weight'][i] == '[125-150)':
        df['weight'] = df['weight'].replace(df['weight'][i],'(125-150)')
    elif df['weight'][i] == '[175-200)':
        df['weight'] = df['weight'].replace(df['weight'][i],'(175-200)')
    elif df['weight'][i] == '[150-175)':
        df['weight'] = df['weight'].replace(df['weight'][i],'(150-175)')
    else:
        pass
```

Replacing wording from ['change'] column containing Ch / No to Yes / No

```python
# Replacing wording from ['change'] column containing Ch / No to Yes / No

for i in range(0,df['change'].shape[0]):
    if df['change'][i]=='No':
        df=df
    elif df['change'][i]=='Ch':
        df['change'] = df['change'].replace(df['change'][i],'Yes')
    else:
        pass
```

```python
# Replacing wording from ['readmitted'] column containing UPPER CASE 'NO' to 'No'

for i in range(0,df['readmitted'].shape[0]):
    if df['readmitted'][i]=='NO':
        df['readmitted'] = df['readmitted'].replace(df['readmitted'][i],'No')
    else:
        pass
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Dropping all data that doesnt not include the important parameter A1Cresult

df = df[-df['A1Cresult'].isnull()].reset_index(drop=True).copy()
```

```python
df.shape
```

```python
# Create a dictionary of columns containing ordered categorical data

ordered_cats = {
    'A1Cresult' : ['Norm','>7','>8'],
    'readmitted' : ['<30','>30','No']
}
```

```python
df['A1Cresult'].info()
```

```python
category = pd.CategoricalDtype(ordered_cats['A1Cresult'], ordered=True)
df['A1Cresult'] = df['A1Cresult'].astype(category)
```

```python
category = pd.CategoricalDtype(ordered_cats['readmitted'], ordered=True)
df['readmitted'] = df['readmitted'].astype(category)
```

```python
crossTab = pd.crosstab(index=df['A1Cresult'],columns=df['readmitted'])
crossTab
```

```python
stats.chi2_contingency(crossTab)
```

# Conclusion

Due to A1Cresults and readmission rates both being classified as categorical data. Chisquared inferential hypothesis was used to see the relationship between the two variable. Because the p-value(0.002) from chisquared is <0.05, we can say that there is a significant difference between the two variables. Hence A1Cresults affects the hospital readmission rates.

The finding is aligned with the conclusion made in the original research article. Further analysis is required to understand what variables affect A1Cresults.
