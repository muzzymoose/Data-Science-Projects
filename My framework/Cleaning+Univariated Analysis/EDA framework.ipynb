# Step 5 : Ask a question about the data
***
> - Try to answer a question you have about the data using a plot or statistics <br>

What are the locations with the fastest roller coaster? (minimum of 10)


```python
df['Location'].value_counts()
```

```python
ax = df.query('Location != "Others"') \
.groupby('Location')['Speed_mph'] \
.agg(['mean','count']) \
.query('count >= 10') \
.sort_values('mean')['mean'] \
.plot(kind='barh', figsize=(12,5), title='Average Coaster Speed by Location')
ax.set_xlabel('Average Coaster Speed')

plt.show()
```
