```python
import os
import json
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
# Configuration
VIDEO_ID = 'P2aqwcKI0u8'
OUTPUT_PATH = '../data/raw/z_channel_apartment.json'

try:
    # 1. Fetch transcript
    api = YouTubeTranscriptApi()
    fetched_objects = api.fetch(VIDEO_ID, languages=['zh-HK', 'zh-TW', 'zh', 'en'])
    
    # 2. Convert to list of dictionaries
    serializable_data = []
    for entry in fetched_objects:
        serializable_data.append({
            'text': entry.text,
            'start': entry.start,
            'duration': entry.duration
        })
    
    # 3. Save to data/raw/
    os.makedirs(os.path.dirname(OUTPUT_PATH), exist_ok=True)
    with open(OUTPUT_PATH, 'w', encoding='utf-8') as f:
        json.dump(serializable_data, f, ensure_ascii=False, indent=4)
        
    print(f"✅ Successfully saved {len(serializable_data)} lines to {OUTPUT_PATH}")

except Exception as e:
    print(f"❌ Extraction failed: {e}")
```
